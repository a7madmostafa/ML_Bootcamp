# Display Text

```python
%%writefile 01-displayText.py

import streamlit as st

st.title ("this is the app title")
st.header("this is the markdown")
st.markdown("this is the header")
st.subheader("this is the subheader")
st.caption("this is the caption")
st.code("x=2021")
st.latex(r''' a+a r^1+a r^2+a r^3 ''')
```

```python
!streamlit run 01-displayText.py
```

# Text Styling

```python
%%writefile 02-textStyles.py

import streamlit as st

# Using Markdown
st.markdown("## This is a markdown header")
st.markdown("### This is a markdown subheader")
st.markdown("*This is a markdown italicized text*")
st.markdown("**This is a markdown bold text**")
st.markdown("~~This is a markdown strikethrough text~~")
st.markdown("> This is a markdown blockquote")
st.markdown("This is a markdown link: [Streamlit](https://streamlit.io/)")
st.markdown("This is a markdown image: ![Streamlit](https://streamlit.io/images/brand/streamlit-mark-color.png)")

# Using HTML
# Coloring and Centering text
st.markdown("<h1 style='text-align: center; color: red;'>Colorful and centered</h1>", unsafe_allow_html=True)
```

```python
!streamlit run 02-textStyles.py
```

# Display progress and status

```python
%%writefile 07-progress.py

import streamlit as st
import time

st.progress(75)
with st.spinner('Wait for it...'):
    time.sleep(10)
    st.balloons()
```

```python
!streamlit run 07-progress.py
```

```python
!streamlit run 08-status.py
```

# Display Graphs

## Matplotlib

```python
%%writefile 10-viz_matplotlib.py

import streamlit as st
import matplotlib.pyplot as plt
import numpy as np

rand=np.random.normal(1, 2, size=20)
fig, ax = plt.subplots()
ax.hist(rand, bins=15)

st.markdown("<h1 style='text-align: center; color: red;'>Matplotlib Figure</h1>", unsafe_allow_html=True)
st.pyplot(fig)
```

```python
!streamlit run 10-viz_matplotlib.py
```

## Plotly

```python
%%writefile 11-viz_plotly.py

import streamlit as st
import plotly.express as px
import numpy as np
import pandas as pd

rand=np.random.normal(1, 2, size=20)
fig = px.histogram(rand, nbins=15)
st.write(' # Plotly Histogram')
st.plotly_chart(fig)

df = pd.DataFrame({'k':[1,2,3,4,5], 'v':[10,20,30,40,50]})
st.write(' # Plotly Bar Chart')
st.plotly_chart(px.bar(df, x='k', y='v'))

st.write(' # Plotly Line Chart')
st.plotly_chart(px.line(df, x='k', y='v'))
```

```python
!streamlit run 11-viz_plotly.py
```

## Easier way to create pages (one file)

```python
%%writefile my_app_2.py

import streamlit as st

def main_page():
    st.markdown("# Main page 🎈")
    st.sidebar.markdown("# Main page 🎈")

def page2():
    st.markdown("# Page 2 ❄️")
    st.sidebar.markdown("# Page 2 ❄️")

def page3():
    st.markdown("# Page 3 🎉")
    st.sidebar.markdown("# Page 3 🎉")

page_names_to_funcs = {
    "Main Page": main_page,
    "Page 2": page2,
    "Page 3": page3,
}

selected_page = st.sidebar.selectbox("Select a page", page_names_to_funcs.keys())
page_names_to_funcs[selected_page]()
```

```python
!streamlit run my_app_2.py
```
