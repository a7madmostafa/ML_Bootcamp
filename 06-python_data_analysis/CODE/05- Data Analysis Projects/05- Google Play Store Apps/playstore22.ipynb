```python
def fix_size(size):
    size = size.lower()
    if 'm' in size:
        x = size[:-1]
        x = float(x)
        return x
    elif 'k' in size:
        x = size[:-1]
        x = float(x)
        x = x / 1000
        return x
    else:
        return np.nan
```

```python
def fix_installs(installs):
    installs = installs.replace('+', '')
    installs = installs.replace(',', '')
    installs = int(installs)
    if installs < 1000 :
        return 'Less than 1k'
    elif installs >= 1000 and installs < 100000:
        return '1k-100k'
    elif installs >= 100000 and installs < 1000000:
        return '100k-1M'
    elif installs >= 1000000 and installs < 10000000:
        return '1M-10M'
    elif installs >= 10000000 and installs < 100000000:
        return '10M-100M'
    else:
        return 'More than 100M'
```

```python
df['Installs'] = df['Installs'].apply(fix_installs)
```

```python
def univariate_numerical(df, col):
    # Subplot
    fig = make_subplots(rows=1, cols=2)
    # Histogram
    fig.add_trace(go.Histogram(x=df[col]), row=1, col=1)
    # Boxplot
    fig.add_trace(go.Box(x=df[col]), row=1, col=2)
    fig.update_layout(title_text=f'{col}')
    fig.update_layout(showlegend=False)
    fig.show()
```

```python
def univariate_categorical(df, col):
    fig = px.bar(df[col].value_counts())
    fig.show()
```

```python
for num_col in numerical_cols:
    univariate_numerical(df, num_col)
```

```python
for cat_col in categorical_cols:
    univariate_categorical(df, cat_col)
```

```python
def bivariate_numerical(df, col1, col2):
    fig = px.scatter(df, x=col1, y=col2, title=f'{col1} vs {col2}')
    fig.show()
```

```python
bivariate_numerical(df, 'Rating', 'Reviews')
```

```python
# Top 10 categories
fig = px.bar(df['Category'].value_counts(normalize=True).sort_values(ascending=False)[:10]*100,
                title='Top 10 Category Percentage',
                labels={'value':'Perentage', 'index':'Category'}, width=800, height=500)
fig.update_layout(showlegend=False)
# Annotations with white font color
annotations = []
for i, v in enumerate(df['Category'].value_counts(normalize=True).sort_values(ascending=False)[:10]*100):
    annotations.append(dict(x=i, y=v-1, text=str(round(v))+'%', font_size=10, showarrow=False))
fig.update_layout(annotations=annotations)
fig.update_annotations(font_color='white')
fig.show()
```

```python
# Bottom 10 categories
fig = px.bar(df['Category'].value_counts().sort_values(ascending=False)[-10:],
                title='Bottom 10 Category Count',
                labels={'value':'Count', 'index':'Category'}, width=800, height=500)
fig.update_layout(showlegend=False)
annotations = []
for i, v in enumerate(df['Category'].value_counts().sort_values(ascending=False)[-10:]):
    annotations.append(dict(x=i, y=v+3, text=str(round(v)), font_size=10, showarrow=False))
fig.update_layout(annotations=annotations)
fig.show()
```
