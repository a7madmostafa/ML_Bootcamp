# Inspecting the dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('zomato.csv', encoding='latin-1')
df.head()
```

```python
codes= pd.read_excel('Country-Code.xlsx')
codes.head()
```

```python
df = df.merge(codes, how='left', on='Country Code')
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()  # Check for duplicate rows
```

```python
df.isna().sum()  # Check for missing values
```

```python
# Drop Missing Values 
df.dropna(inplace=True)
```

```python
# Drop Unnecessary Columns
df.drop(['Restaurant ID','Country Code', 'Address', 'Locality Verbose', 'Locality', 'Longitude', 'Latitude'], axis=1, inplace=True)
```

```python
# Rearrange Columns
df = df[['Restaurant Name', 'City', 'Country', 'Cuisines', 'Average Cost for two', 'Currency', 'Has Table booking',
         'Has Online delivery', 'Is delivering now', 'Switch to order menu', 'Price range', 'Aggregate rating',
         'Rating color', 'Rating text', 'Votes']]
```

```python
# from ydata_profiling import ProfileReport
# profile = ProfileReport(df, title="Zomato Restaurants Data Profiling Report", explorative=True)
# profile.to_file("zomato_profiling_report.html")
```

```python
# Remove duplicate rows (Appeared after ID column was dropped)
df.drop_duplicates(inplace=True)  # Remove duplicate rows
```

```python
# Drop unnecessary columns
# 'Switch to order menu' has constant values (False)
# 'Rating color' is redundant with 'Rating text'
df.drop(['Switch to order menu', 'Rating color'], axis=1, inplace=True)
```

```python
# Rename columns for consistency
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.describe(include='O')  # Get descriptive statistics for categorical columns
```

```python
df['price_range'].value_counts()  # Check unique values in 'Price range'
```

```python
df['price_range'] = df['price_range'].astype('category')  
```

```python
# Numercial columns 
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Categorical columns
cat_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
```

```python
df['average_cost_for_two'].quantile([0.25, 0.5, 0.75, 0.9, 0.95, 0.995, 0.999])
```

```python
df = df[df['average_cost_for_two'] > 0]  # Remove rows with zero cost
df = df[df['average_cost_for_two'] < df['average_cost_for_two'].quantile(0.995) ]  # Remove outliers
```

```python
df.shape
```

# Univariate Analysis 

```python
def explore_num_col(df, col):
    fig = px.histogram(df, x=col, marginal='box', title=f'Distribution of {col}', width=900)
    fig.update_layout(xaxis_title=col, yaxis_title='Count')
    fig.show()

def explore_cat_col(df, col):
    fig = px.histogram(df, x=col, title=f'Distribution of {col}', category_orders={col: df[col].value_counts().index}, width=900)
    fig.update_layout(xaxis_title=col, yaxis_title='Count')
    fig.show()
```

```python
for col in num_cols:
    explore_num_col(df, col)
```

```python
for col in cat_cols:
    explore_cat_col(df, col)
```

* Insights 
  * Most Numerical Columns are skewed to the right
  * Some restaurants having very high costs (Average Cost for two)
  * Many restaurants have zero aggregate rating
  * A few restaurants have very high number of votes
  * Most restaurants in India
  * Most restaurants have online delivery and table booking options
  * Most restaurants have a price range of 1 or 2
  * Almost all restaurants have IS delivering now as 'No'
  * Excellent and Poor ratings are the least common while Average is the most common and Some restaurants have no ratings at all


# 📌 Market & Expansion Strategy

## Which countries and cities have the highest concentration of restaurants?

```python
df.columns
```

```python
df.country.value_counts() 
```

```python
df.city.value_counts()
```

```python
print(f'Top Country: {df["country"].value_counts().idxmax()}')
print(f'Top City: {df["city"].value_counts().idxmax()}')
```

```python
# Top 3 countries with most restaurants
top_countries = df['country'].value_counts().nlargest(3)
top_countries
```

```python
other_countries_count = df.country.size - top_countries.sum() 
other_countries_count
```

```python
top_countries.loc['Others'] = other_countries_count
top_countries
```

```python
# Top 3 cities with most restaurants
top_cities = df['city'].value_counts().nlargest(3)
other_cities_count = df.city.size - top_cities.sum()
top_cities.loc['Others'] = other_cities_count
top_cities
```

```python
px.pie(top_countries, values=top_countries.values, names=top_countries.index, width=900, title='Top countries with most restaurants')
```

```python
px.pie(top_cities, values=top_cities.values, names=top_cities.index, title='Top Cities with Most Restaurants', width=900)
```

## What are the most popular cuisines globally, and how do they vary by country?

```python
df.cuisines.value_counts().nlargest(10)
```

```python
popular_cuisines = (
    df.groupby('country')['cuisines']
      .apply(lambda x: x.value_counts().head(1))  
      .sort_values(ascending=False).reset_index()
)

popular_cuisines.columns = ['country', 'most_popular_cuisine', 'count']

popular_cuisines
```

```python
px.bar(df.cuisines.value_counts().nlargest(10),
       x=df.cuisines.value_counts().nlargest(10).index,
       y=df.cuisines.value_counts().nlargest(10).values,
       width=900, title='Top 10 Cuisines')
```

## What are the most popular restaurants?

```python
popular_restaurants = df.groupby('restaurant_name').size().nlargest(10)
popular_restaurants
```

```python
px.bar(popular_restaurants,
       x=popular_restaurants.index,
       y=popular_restaurants.values,
       width=900, title='Top 10 Restaurants')
```

## Which restaurants have the highest number of votes?

```python
top_restaurants = df.nlargest(10, 'votes')[['restaurant_name', 'country', 'city',  'cuisines', 'votes']].reset_index(drop=True)
top_restaurants
```

```python
top_restaurants_ordered = top_restaurants.groupby('restaurant_name')['votes'].sum().sort_values(ascending=False).index.to_list()
top_restaurants_ordered
```

```python
px.bar(top_restaurants, x='restaurant_name', y='votes', color='city',
       title='Top 10 Restaurants by Votes', width=900, height=600, color_discrete_sequence=px.colors.qualitative.Dark2,
       category_orders={'restaurant_name': top_restaurants_ordered}
)
```

```python
cuisines = df['Cuisines'].dropna().str.split(',').explode().str.strip()
top_cuisines = cuisines.value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette="crest")
plt.title("Top 10 Most Popular Cuisines")
plt.xlabel("Number of Restaurants")
plt.show()
```
