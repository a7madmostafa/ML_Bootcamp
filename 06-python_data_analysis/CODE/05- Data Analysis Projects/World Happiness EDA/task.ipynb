# 🌍 World Happiness 2021 Dashboard — Task Requirements

## 🎯 Objective
Build an **interactive Streamlit dashboard** using the *World Happiness Report 2021* dataset.  
The goal is to explore **what drives happiness across countries and regions** and provide insights for policymakers, NGOs, and analysts.

---

## 📌 Dashboard Requirements

### 1️⃣ Overview
- Display the **average global happiness score**.  
- Show **Top 10 happiest** and **Bottom 10 countries**.  

### 2️⃣ Key Drivers
- Correlation heatmap between **Happiness** and main features:  
  `GDP, Social support, Life Expectancy, Freedom, Generosity, Corruption`.

### 3️⃣ Regional Insights
- Compare **average happiness by region** with an interactive bar chart.  

### 4️⃣ Country Comparison
- Allow selection of **2–3 countries** for side-by-side comparison.  
- Visualize **Happiness vs GDP** to spot **overperformers and underperformers**.

### 5️⃣ Income Groups
- Group countries into **Low, Middle, High income (by GDP quantiles)**.  
- Compare average **Happiness across income groups**.

### 6️⃣ Conclusions
- Summarize main findings directly in the dashboard (e.g., GDP, Social Support, and Life Expectancy are strongest drivers).

---

## ✅ Deliverables
- 📂 **GitHub Repository** with:
  - Cleaned dataset (`world_happiness_2021.csv`)
  - Streamlit app code (`world_happiness_app.py`)
  - README with setup instructions  

- 🌐 **Deployed Streamlit App** link to share and interact with the dashboard.
