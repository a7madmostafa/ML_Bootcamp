### enumerate

```python
for i, num in enumerate([10,20,30]):
    print(i)
```

```python
for i,letter in enumerate('Hello'):
    print("index {}: {}".format(i,letter))
```

```python
dic={}
for i,letter in enumerate('Hello'):
    dic[letter] = i

dic
```

```python
list(enumerate('Hello'))
```
