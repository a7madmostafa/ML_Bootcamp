# Importing  libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.tree import DecisionTreeRegressor
```

# Random Data

```python
# Generate a random dataset
np.random.seed(42)
X_quad = np.random.rand(200, 1) - 0.5  # a single random input feature
y_quad = X_quad ** 2 + 0.025 * np.random.randn(200, 1)

# Plot the dataset
plt.scatter(X_quad, y_quad, s=25, c='blue', edgecolor='k', label='Dataset')
plt.show()
```

# Decision Tree Regressor

```python
tree_reg_2 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg_2.fit(X_quad, y_quad)

print("Accuracyfor Max_depth = 2: ", tree_reg_2.score(X_quad, y_quad))

# Visualize the Decision Tree 
from sklearn.tree import plot_tree
plt.figure(figsize=(15,7))
plot_tree(tree_reg_2, feature_names=['x'], filled=True, rounded=True)
plt.show()
```

```python
tree_reg_3 = DecisionTreeRegressor(max_depth=3, random_state=42)
tree_reg_3.fit(X_quad, y_quad)

print("Accuracyfor Max_depth = 3: ", tree_reg_3.score(X_quad, y_quad))

# Visualize the Decision Tree 
from sklearn.tree import plot_tree
plt.figure(figsize=(15,7))
plot_tree(tree_reg_3, feature_names=['x'], filled=True, rounded=True, fontsize=14)
plt.show()
```

```python
def plot_regression_predictions(tree_reg, X, y, axes=[-0.5, 0.5, -0.05, 0.25]):
    x1 = np.linspace(axes[0], axes[1], 500).reshape(-1, 1)
    y_pred = tree_reg.predict(x1)
    plt.axis(axes)
    plt.xlabel("$x_1$")
    plt.plot(X, y, "b.")
    plt.plot(x1, y_pred, "r.-", linewidth=2)
```

```python
plt.figure(figsize=(15,5))
# Subplot 1
plt.subplot(121)
plot_regression_predictions(tree_reg_2, X_quad, y_quad)
plt.title("max_depth=2", fontsize=14)

# Subplot 2
plt.subplot(122)
plot_regression_predictions(tree_reg_3, X_quad, y_quad)
plt.title("max_depth=3", fontsize=14)

plt.show()
```

```python
tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(X_quad, y_quad)

print("Accuracy ", tree_reg.score(X_quad, y_quad))

# Visualize the Decision Tree 
from sklearn.tree import plot_tree
plt.figure(figsize=(15,7))
plot_tree(tree_reg, feature_names=['x'], filled=True, rounded=True, fontsize=14)
plt.show()
```
